```python
# create a list of the signs to look for
signs = ['aries', 
         'taurus', 
         'gemini', 
         'cancer', 
         'leo', 
         'virgo', 
         'libra', 
         'scorpio', 
         'sagittarius', 
         'capricorn', 
         'aquarius', 
         'pisces']

# Create a search pattern that can be used in a regular expression
pattern = '|'.join(signs)
pattern
```

```python
# import the regular expressions package
import re

# Define a function that finds the pattern in a string
def clean_sign(sign_text):
    # In case the inputs are messy, check to see the input is a string type
    if type(sign_text) == str:
        # Check to see if there's a match
        m = re.search(pattern, sign_text)
        # In case there is no match, just return ''
        if m is None:
            return ''
        else:
            return m.group(0)
    else:
        return "The input is not a string"
```

```python
clean_sign('acquarius virgo')
```

```python
df_signs['clean_signs'] = df_signs['sign'].apply(clean_sign)
```
